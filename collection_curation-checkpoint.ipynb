# merging

```python
import pandas as pd
```

```python
file1 = open("part_0.csv" , "a")
file2 = open("part_1.csv" , "r" )

for line in file2:
    file1.write(line)
    
file1.close()
file2.close()
```

# dataframe

```python
df = pd.read_csv("part_0.csv")
print(df.info())
```

```python
df.head()
```

# outcome ratio by gender

```python
import numpy as np
import matplotlib.pyplot as plt

data = df.groupby("Gender")["Outcome_indexed"].sum()

pie, ax = plt.subplots(figsize=[10,6])
data.plot.pie(autopct="%.1f%%", pctdistance=0.8)
plt.title("Outcome ratio by Gender", fontsize = 18);
```

# outcome ratio by outbreak

```python
import numpy as np
import matplotlib.pyplot as plt

data = df.groupby("Outbreak")["Outcome_indexed"].sum()

pie, ax = plt.subplots(figsize=[20,12])
data.plot.pie(autopct="%.1f%%", pctdistance=0.8)
plt.title("Outbreak ratio by Gender", fontsize = 18);
```

# outcome by age group

```python
import seaborn as sns

data2 = df.groupby("Age Group")["Outcome_indexed"].sum()

f, ax = plt.subplots(figsize = (20, 10))

ax = data2.plot.bar(x = 'Age', y = 'Fatality')

sns.set_context("poster")

plt.xlabel('Age Group', fontsize = 14, weight = 'bold')
plt.ylabel('Fatality', fontsize = 14, weight = 'bold')

f.savefig('bar') 
```

# outcome by intubated

```python
import numpy as np
import matplotlib.pyplot as plt

data = df.groupby("Intubated")["Outcome_indexed"].sum()

pie, ax = plt.subplots(figsize=[20,12])
data.plot.pie(autopct="%.1f%%", pctdistance=0.5)

plt.title("Fatality by being in Intubated", fontsize = 20);
```

# outcome by source of infection

```python
data = df.groupby("Source of Infection")["Outcome_indexed"].sum()

f, ax = plt.subplots(figsize = (30, 15))
ax = data.plot.bar(x = 'Source of Infection', y = 'Fatality')
plt.palette = ("Red")
plt.ylabel("Outcome")
plt.title("Total Outcome by Source of Infection" , fontsize = 20)
```
